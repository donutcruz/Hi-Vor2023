```python
import numpy as np
import matplotlib.pyplot as plt
import urllib.request

url = 'https://cs.hi.is/python/karfa.txt'
response = urllib.request.urlopen(url)
data = np.loadtxt(response, unpack=True)

haed, thyngd = data[1], data[2]

# Teiknum punktarit af gögnunum
plt.scatter(haed, thyngd)
plt.title('Hæð og þyngd')
plt.xlabel('Hæð (fet)')
plt.ylabel('Þyngd (pund)')

# Finnum jöfnu bestu línu
m, b = np.polyfit(haed, thyngd, 1)

# Teiknum bestu línu ásamt punktaritinu
plt.plot(haed, m*haed + b, color='red')

plt.show()
```

## Lasun 27.3


```python
import numpy as np
import matplotlib.pyplot as plt
import urllib.request

url = 'https://cs.hi.is/python/karfa.txt'
response = urllib.request.urlopen(url)
data = np.loadtxt(response, unpack=True)

haed, thyngd = data[1], data[2]

# Teiknum punktarit af gögnunum
plt.scatter(haed, thyngd)
plt.title('Hæð og þyngd')
plt.xlabel('Hæð (fet)')
plt.ylabel('Þyngd (pund)')

# Finnum jöfnu bestu parabólunnar
p = np.polyfit(haed, thyngd, 2)

# Teiknum bestu parabóluna ásamt punktaritinu
x = np.linspace(min(haed), max(haed), 100)
y = p[0]*x**2 + p[1]*x + p[2]
plt.plot(x, y, color='red')

plt.show()
```

```python
import numpy as np
import urllib.request

url = 'https://cs.hi.is/python/karfa.txt'
response = urllib.request.urlopen(url)
data = np.loadtxt(response, unpack=True)

aldur = data[0]
med_aldur = np.median(aldur)

print("Miðtala aldurs körfuboltamanna: ", med_aldur)

yngri = data[:, aldur < med_aldur]
eldri = data[:, aldur >= med_aldur]

med_haed_yngri = np.median(yngri[1])
med_haed_eldri = np.median(eldri[1])

print("Meðalhæð yngri hópsins: ", med_haed_yngri)
print("Meðalhæð eldri hópsins: ", med_haed_eldri)
```

# 34.Mannfjöldaspá fyrir Ísland 




## Lausn 34.1


```python
import urllib.request
import numpy as np

# Sækja gögn um aldurshátt íslenskra borgara
url = 'https://cs.hi.is/python/aldursdreifing.txt'
response = urllib.request.urlopen(url)
data = np.genfromtxt(response, skip_header=1, usecols=(1, 2, 3, 4), unpack=True)

# Fjöldi ára sem við viljum spá fyrir um
ar = np.arange(2020, 2076)

# Fjöldi fæðinga á hverju ári
faedingar = data[0]

# Fjöldi dánna á hverju ári
danir = data[1]

# Meðalævilíkur á hverju ári
lifun = data[2]

# Heildarfjöldi íbúa á hvert ár
N = np.zeros_like(ar, dtype=np.float64)
N[0] = 364260

# Reikna heildarfjölda íbúa með formúlunni fyrir hverju ári
for i in range(1, len(ar)):
    N[i] = N[i-1] + faedingar[i-1] - danir[i-1]
    N[i] = N[i] * (1 + (lifun[i-1]/1000))

# Prenta töflu yfir heildarfjölda á 5 ára fresti
print('Ár\tHeildarfjöldi íbúa')
print('----------------------')
for i in range(0, len(ar), 5):
    print(f'{ar[i]}\t{int(N[i])}')
```
