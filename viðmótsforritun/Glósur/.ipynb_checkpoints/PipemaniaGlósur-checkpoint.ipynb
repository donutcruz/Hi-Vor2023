### Skref 2
Búðu til nýtt IntelliJ project og búðu  til controller klasa t.d. **PipemaniaController** og aðalklasa **PipemaniaApplication** og settu klasana í **vidmot pakka**. Hér er best að nota Refactor í IntelliJ (hægri smella á nafn skráarinnar / Refactor / Rename) 

___

Hérna lýtur project-structure út hjá mér - Svo er annað package þarna sem ég bætti við til að innihalda myndum af pípunum.

- 5x Classar 
- **Vidmot** inniheldur *Controller* og *PipApplication*
- **vinnsla** inniheldur *Pipa*, *Pipemania*, og *Pipukista*
- **Myndir** Fer svo inn í hvernig ég bjó það til á eftir.



Þegar við höfum bætt þessu inn í project structure - vertu viss um að fara yfir þessa hluti. 

- Inn í **pipemania-view.fxml** er ekki alveg pottþétt búið að declare-a Controller? Þar sem Borderpane er efst hjá mér í Sccenebuilder lýtur þetta út svona. Ef Controller er ekki declare-að fáiði risa villu þegar Application er keyrt.





## Skref 3.

Í **SceneBuilder** skaltu búa til notendaviðmótið fyrir skissuna sem þú gerðir.
Ekki setja neina atburðahandlera eða tilviksbreytur strax. **Settu nafnið á .fxml skránni í PipemaniaApplication klasann.**



Ég setti þetta svona upp. Hérna er ekki komið neinar atburðahandlera eða tilviksbreytur.

- Efst (Base-ið mitt) er **Borderpane** og allt hitt er inni því.
- Hægra meginn er *GridPane* og inní því er *ImageView* (Hérna koma random pípurnar)
- Stóra griddið er **leikborðið** Gridpane 5x5 og inní því eru buttons. (Hér velur leikmaður reiti til að setja nýju pípurnar 
- Neðst er grid líka 3x1 og inní því eru 2x label *Stig* (bara texti), og *StigX* (í StigX kemur framm hversu mörg stig þú ertbúin að fá, og 1x Button *Flæði* (Þetta er button sem að þú ýtir til að AthFlaedi)

## Skref 4.

Settu **fxVeljaReit** handler á **hnappana** á leikborðinu. Skrifaðu einfalda prentsetningu á staðalúttak til að fullvissa þig um að handlerinn virki.



- Fannst langt þægilegast að ýta á cmd + A - til að velja alla Buttons inní Grid 5x5 
- Settu svo **VeljaReit** í **OnAction** undir Code neðst hægra meginn í Scenebuilder
- Ebba biður okkur um að setja einfaldan staðútakk, ég bíð aðeins með það þangað tal ég er búin að setja allt upp og geri það svo í lokinn. Hægt er að setja System.out.println("Reit var valið"); fyrir **VeljaReit**


## Skref 5.

 Settu **fxAthFlaedir** handler á hnappinn sem athugar hvort vatn flæðir frá upphafspípu og í gegnum pípulögnina.
Skrifaðu einfalda prentsetningu á staðalúttak til að fullvissa þig um að handlerinn virki. 

- Þetta er í raun alveg eins og skref 4. nema **Flæði button** inni scenebuilder er **OnAction** *AthFlaedi*
